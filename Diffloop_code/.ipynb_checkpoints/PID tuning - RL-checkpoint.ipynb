```python
import numpy as np
import scipy
import torch
from matplotlib import pyplot as plt
import torch.optim as optim
import torch.nn.functional as F
from torch.autograd import Variable 
import control
from control import matlab
from matplotlib import cm

np.random.seed(0)
dev = "cpu"  
device = torch.device(dev) 
```

```python
sysA = 1.010
#sysA = 0.99
sysB = 0.125
sysC = 0.1608
sysD = 0

kp = 4.0
ki = 16.0
```

```python
x = 0

ref = np.ones((1500,))
ref[200:400] = 2;
ref[400:600] = 0;
ref[600:1000] = -2.1;
ref[1000:1400] = 2.0;
ref[1400:1500] = 3.0;

ysat = np.zeros((1500,))
uvalssat = np.zeros((1500,))
xcont = 0;
u = 0;
for i in range(1500):
   ysat[i] = sysC*x+sysD*u;
   xcont = xcont+(ref[i]-ysat[i]);
   u = 0.01*ki*xcont+kp*(ref[i]-ysat[i]);
   if(u>4):
     u=4;
   elif(u<-4):
     u=-4;  
   uvalssat[i] = u;
   x = sysA*x+sysB*u;

ynosat = np.zeros((1500,))
uvalsnosat = np.zeros((1500,))
xcont = 0;
u = 0;
for i in range(1500):
   ynosat[i] = sysC*x+sysD*u;
   xcont = xcont+(ref[i]-ynosat[i]);
   u = 0.01*ki*xcont+kp*(ref[i]-ynosat[i]); 
   uvalsnosat[i] = u;
   x = sysA*x+sysB*u;
```

```python
plt.figure
plt.plot(0.01*np.arange(1500),ynosat)
plt.plot(0.01*np.arange(1500),ysat)
plt.xlabel("Time (s)")
plt.ylabel("Output")
plt.show()

plt.figure
plt.plot(0.01*np.arange(1500),uvalsnosat)
plt.plot(0.01*np.arange(1500),uvalssat)
plt.xlabel("Time (s)")
plt.ylabel("Input")
plt.show()
```

```python
# Use policy gradient to tune kp,ki
ysat = np.zeros((1500,))
uvalssat = np.zeros((1500,))

kppg = kp
kipg = ki
nupdates = 100
nepisodes = 128
updatecosts = np.zeros((nupdates,))
for update in range(nupdates):
  kptvals = np.zeros((nepisodes,))
  kitvals = np.zeros((nepisodes,))
  costs = np.zeros((nepisodes,))
  kpder = np.zeros((nepisodes,))
  kider = np.zeros((nepisodes,))  
  for episode in range(nepisodes):
    ref = np.ones((1500,))  
    #ref[200:400] = np.random.randint(-3,3);
    #ref[400:600] = np.random.randint(-3,3)
    #ref[600:1000] = np.random.randint(-3,3)
    #ref[1000:1400] = np.random.randint(-3,3)
    #ref[1400:1500] = np.random.randint(-3,3)
    ref[200:400] = 2;
    ref[400:600] = 0;
    ref[600:1000] = -2.1;
    ref[1000:1400] = 2.0; 
    ref[1400:1500] = 3.0;
    sigma = 1                          
    
    kpt = sigma*(np.random.rand()-0.5)+kppg
    kit = sigma*(np.random.rand()-0.5)+kipg
    kptvals[episode] = kpt
    kitvals[episode] = kit 
   
    xcont = 0;
    u = 0;
    x = 0;                                 
    for t in range(1500):
      ysat[t] = sysC*x+sysD*u;
      xcont = xcont+(ref[t]-ysat[t]);
      u = 0.01*kit*xcont+kpt*(ref[t]-ysat[t]); 
      uvalssat[t] = u;
      if(u>4):
        u=4
      elif(u<-4):
        u=-4                            
      x = sysA*x+sysB*u;                                
      costs[episode] += (ref[t]-ysat[t])**2
      #kpder[episode] += -(kpt-kppg)*(ref[t]-ysat[t])/(2*sigma)
      #kider[episode] += -(kit-kipg)*0.01*xcont/(2*sigma)
      #kpder[episode] += -(kpt-kppg)*(ref[i]-ysat[i])/(2*sigma)
      #kider[episode] += -(kit-kipg)*0.01*xcont/(2*sigma)
    #print("Episode cost: ",costs[episode])    
  # Update kp,ki
  lr = 1
  updatecosts[update] = np.mean(costs)
  print("Update cost",updatecosts[update])
  kppg = kppg-lr*np.clip(np.mean((kptvals-kppg)*(costs[episode]-np.mean(costs))),-1000,1000)
  kipg = kipg-lr*np.clip(np.mean((kitvals-kipg)*(costs[episode]-np.mean(costs))),-1000,1000)
  #print(kppg)
  #print(kipg)  
  #if(kppg<0):
  #  kppg = 0
  #if(kipg<0):
  #  kipg = 0                                 
                                     
```
