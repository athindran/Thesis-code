```python
systA = torch.tensor(sysA,requires_grad=False)
systB = torch.tensor(sysB,requires_grad=False)
systC = torch.tensor(sysC,requires_grad=False)
systD = torch.tensor(sysD,requires_grad=False)

refc1 = np.zeros((1000,))
refc1[0:250] = 4
refc1[250:400] = -3
refc1[400:700] = 2
refc1[700:900] = 1

refc1t = torch.tensor(refc1,requires_grad=False)
#refc2t = torch.tensor(np.sin(np.arange(0,1000)*100),requires_grad=False)
#reftott = refc1t+refc2t
refc2t = torch.tensor(refc1,requires_grad=False)

kp1 = torch.tensor(kp,requires_grad=True)
kp2 = torch.tensor(kp,requires_grad=True)

ki1 = torch.tensor(ki,requires_grad=True)
ki2 = torch.tensor(ki,requires_grad=True)

kd1 = torch.tensor(kd,requires_grad=True)
kd2 = torch.tensor(kd,requires_grad=True)

sat1 = torch.tensor([300.0],requires_grad=True)
sat2 = torch.tensor([300.0],requires_grad=True)

Tf = torch.tensor([0.01],requires_grad=False)
nc1 = 0
nc2 = 0
for episodes in range(5001):
  xcont1 = torch.tensor([0.0],requires_grad=True)
  xcont2 = torch.tensor([0.0],requires_grad=True)
  cost1 = torch.tensor([0.0],requires_grad=True)
  cost2 = torch.tensor([0.0],requires_grad=True)
  D1  = torch.tensor([0.0],requires_grad=False)
  D2  = torch.tensor([0.0],requires_grad=False)

  ad1 = torch.div(Tf,(Tf+0.01))
  bd1 = torch.div(kd1,(Tf+0.01))

  ad2 = torch.div(Tf,(Tf+0.01))
  bd2 = torch.div(kd2,(Tf+0.01))
 
  noise = 0.05*torch.randn(1000,1)
    
  x = torch.tensor([[0.0],[0.0],[0.0]],requires_grad=True)  
  u = torch.tensor([0.0],requires_grad=True)
  yt = torch.tensor([0.0],requires_grad=True)  
  #print(kp1,ki1,kd1)
  for i in range(1000):
    yold = yt  
    yt = torch.clamp(torch.matmul(systC.float(),x)+noise[i],-10,10);
    #print(yt)
    xcont1 = xcont1+(refc1t[i]-yt)
    if(i>0):
      D1 = ad1*D1-bd1*(yt-yold)  
      #u = sat1*torch.tanh((0.01*ki1*xcont1+kp1*(refc1t[i]-yt)+D1)/sat1)
      u = (0.01*ki1*xcont1+kp1*(refc1t[i]-yt)+D1)
      u = torch.clamp(u,-300.0,300.0)
    x = torch.matmul(systA.float(),x)+systB.float()*u
    cost1 = cost1+(refc1t[i]-yt)**2 
  
  x = torch.tensor([[0.0],[0.0],[0.0]],requires_grad=True)  
  u = torch.tensor([0.0],requires_grad=True)
  yt = torch.tensor([0.0],requires_grad=True)

  for i in range(1000):
    yold = yt  
    yt = torch.clamp(torch.matmul(systC.float(),x)+noise[i],-10,10)
    xcont2 = xcont2+(refc2t[i]-yt)
    if(i>0): 
      D2 = ad2*D2-bd2*(yt-yold)
      #u = sat2*torch.tanh((0.01*ki2*xcont2+kp2*(refc2t[i]-yt)+D2)/sat2)
      u = (0.01*ki2*xcont1+kp2*(refc2t[i]-yt)+D2)
      u = torch.clamp(u,-300.0,300.0)
      #u = torch.clamp((u1+u2)/2.0,-250.0,250.0)
    x = torch.matmul(systA.float(),x)+systB.float()*u;
    cost2 = cost2+(refc2t[i]-yt)**2
  cost = cost1+cost2
  if(episodes%100==0):
    print(episodes,cost1,cost2)
  if(episodes%500==0):
    if(episodes==0):
      optimizer = torch.optim.Adam((kp1,ki1,kd1,sat1),lr=0.1)  
    elif(cost1<cost2):
      print("Controller 1 wins")  
      nc1 = nc1+1
      kp1final = kp1
      ki1final = ki1
      kd1final = kd1  
      sat1final = sat1
      kp2 = torch.tensor(10*kp2*torch.rand(1)+0.1*kp2,requires_grad=True)
      ki2 = torch.tensor(10*ki2*torch.rand(1)+0.1*ki2,requires_grad=True)
      kd2 = torch.tensor(10*kd2*torch.rand(1)+0.1*kd2,requires_grad=True)
      sat2 = torch.tensor(1.5*sat2*torch.rand(1)+0.5*sat2,requires_grad=True)
      optimizer = torch.optim.Adam((kp2,ki2,kd2,sat2), lr=0.1)
    else:
      print("Controller 2 wins")
      nc2 = nc2+1
      kp2final = kp2
      ki2final = ki2
      kd2final = kd2
      sat2final = sat2  
      kp1 = torch.tensor(10*kp1*torch.rand(1)+0.1*kp1,requires_grad=True)
      ki1 = torch.tensor(10*ki1*torch.rand(1)+0.1*ki1,requires_grad=True)
      kd1 = torch.tensor(10*kd1*torch.rand(1)+0.1*kd1,requires_grad=True)
      sat1 = torch.tensor(1.5*sat1*torch.rand(1)+0.5*sat1,requires_grad=True)
      optimizer = torch.optim.Adam((kp1,ki1,kd1,sat1), lr=0.1)     
    if(abs(nc1-nc2)>=7):
      break
  else:
    optimizer.zero_grad()
    #optimizer2.zero_grad()
    cost.backward()
    #optimizer1.step()
    optimizer.step()
```

```python
# Use policy gradient with previous controller cost as baseline
x = np.array([[0],[0],[0]])
ref = refc1
kppg = 50.0
kipg = 0.1
kdpg = 10.0
satpg = 350
Tf = 0.01

nupdates = 50
nepisodes = 128
updatecosts = np.zeros((nupdates,))

for update in range(nupdates):
  kptvals = np.zeros((nepisodes,))
  kitvals = np.zeros((nepisodes,))
  kdtvals = np.zeros((nepisodes,))
  sattvals = np.zeros((nepisodes,))
  costs = np.zeros((nepisodes,))  
  for episode in range(nepisodes):
    ypg = np.zeros((1000,))
    uvalspg = np.zeros((1000,))
    sigma = 1            
    
    kpt = sigma*(np.random.rand()-0.5)+kppg
    kit = sigma*(np.random.rand()-0.5)+kipg
    kdt = sigma*(np.random.rand()-0.5)+kdpg
    satt = 10*sigma*(np.random.rand()-0.5)+satpg
    kptvals[episode] = kpt
    kitvals[episode] = kit
    kdtvals[episode] = kdt
    sattvals[episode] = satt
    #print(kpt,kit,kdt,satt)
    xcont = 0
    u = 0
    D = 0
    ad = Tf/(Tf+0.01)
    bd = kdt/(Tf+0.01)
    x = np.array([[0],[0],[0]])
    noise = 0.05*np.random.randn(1000,1)

    for i in range(1000):
      #print(u) 
      ypg[i] = np.dot(sysC,x)+np.dot(sysD,u)+noise[i];
      #print(ysat[i])
      xcont = xcont+(ref[i]-ypg[i]);
      if(i>0):
        D = ad*D-bd*(ypg[i]-ypg[i-1]);  
        u = satt*np.tanh((0.01*kit*xcont+kpt*(ref[i]-ypg[i])+D)/satt)
      #if(u>=330):
      #  u = 330
      #elif(u<=-330):
      #  u = -330
      uvalssat[i] = u
      x = np.dot(sysA,x)+np.dot(sysB,u)
      costs[episode] += (ref[i]-ypg[i])**2

    costs[episode] = min(costs[episode],10000)
  lr = 0.1
  updatecosts[update] = np.mean(costs)
  print("Update cost",updatecosts[update])
  kppg = kppg-lr*np.mean((kptvals-kppg)*(costs[episode]-np.mean(costs)))
  kipg = kipg-lr*np.mean((kitvals-kipg)*(costs[episode]-np.mean(costs)))
  kdpg = kdpg-lr*np.mean((kdtvals-kdpg)*(costs[episode]-np.mean(costs)))
  satpg = satpg-lr*np.mean((sattvals-satpg)*(costs[episode]-np.mean(costs)))
  print(kppg,kipg,kdpg,satpg)
```

```python
# Robust Optimization
systA = torch.tensor(sysA,requires_grad=False)
systB = torch.tensor(sysB,requires_grad=False)
systC = torch.tensor(sysC,requires_grad=False)
systD = torch.tensor(sysD,requires_grad=False)


advA = torch.randn(3,3,requires_grad=True)
advB = torch.randn(3,1,requires_grad=True)

kpt = torch.tensor(kp,requires_grad=True)
kit = torch.tensor(ki,requires_grad=True)
kdt = torch.tensor(kd,requires_grad=True)
Tft = torch.tensor(Tf,requires_grad=True)
satt = torch.tensor([350.0],requires_grad=True)
optimizer = torch.optim.Adam((kpt,kit,kdt,satt), lr=0.1)
optimadv = torch.optim.Adam((advA,advB), lr=0.01)
#Tfoptimizer = torch.optim.Adam([Tft],lr=0.0001)
advpower = 0.01
actsat = 350.0

strengthenadversary = False
nc = 0
na = 0
for episodes in range(5001):
  #kpt = F.relu(kpt)
  #kit = F.relu(kit)
  reft = torch.tensor(refs[5+episodes%5,:])  
  xcont = torch.tensor([0.0],requires_grad=True)
  x = torch.tensor([[0.0],[0.0],[0.0]],requires_grad=True)  
  u = torch.tensor([0.0],requires_grad=True)
  cost = torch.tensor([0.0],requires_grad=True)
  advcost = torch.tensor([0.0],requires_grad=True)
  yt = torch.tensor([0.0],requires_grad=True)
  D  = torch.tensor([0.0],requires_grad=False)
  #torch.clamp(Tft,0.1,0.1)
  ad = torch.div(Tft,(Tft+0.01))
  bd = torch.div(kdt,(Tft+0.01))
  noise = 0.05*torch.randn(1000,1)  
  for t in range(1000):
    yold = yt
    yt = torch.clamp(torch.matmul(systC.float(),x)+noise[t],-10.0,10.0)
    #error = torch.tensor(reft[t]-yt,requires_grad=False)
    xcont = xcont+(reft[t]-yt)
    if(t>0):
       #deriv = torch.tensor(yt-yold,requires_grad=False) 
       D = ad*D-bd*(yt-yold);  
       u = satt*torch.tanh((0.1*kit*xcont+kpt*(reft[t]-yt)+D)/satt)
    u = torch.clamp(u,-450.0,450.0)
    adv = torch.matmul(advA.float(),x)+advB.float()*u
    x = torch.matmul(systA.float(),x)+systB.float()*u+torch.clamp(adv,-1*advpower,advpower)
    cost = cost+torch.log(torch.abs(reft[t]-yt)+0.001)
    advcost = advcost-yt**2
  if(episodes%300==0):
    #print(kpt,kdt,kit,Tft)
    print(episodes,cost,advcost)
    if(episodes==0):
       prevcost = cost
    if(episodes!=0):
      if(not strengthenadversary and cost<prevcost):
        prevcost = cost
        print("Controller progress")  
        #advpower = advpower*2
        print(kpt,kit,kdt,satt)
        strengthenadversary = True
        nc = nc+1
      elif(strengthenadversary and advcost<prevadvcost):
        print("Adversary progress")  
        #advpower = advpower/2
        prevadvcost = advcost
        #kpt = torch.tensor(1.5*kpt*torch.rand(1)+0.5*kpt,requires_grad=True)
        #kit = torch.tensor(1.5*kit*torch.rand(1)+0.5*kit,requires_grad=True)
        #kdt = torch.tensor(1.5*kdt*torch.rand(1)+0.5*kdt,requires_grad=True)
        #satt = torch.tensor(1.5*satt*torch.rand(1)+0.5*satt,requires_grad=True)
        optimizer = torch.optim.Adam((kpt,kit,kdt,satt), lr=0.1)
        strengthenadversary = False
        na = na+1
      elif(strengthenadversary):
        print("Controller won, increasing adversary power")
        advpower = advpower*2;
    prevcost = cost
    prevadvcost = advcost
  if not strengthenadversary:      
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
  else:
    optimadv.zero_grad()
    advcost.backward()
    optimadv.step()
```
