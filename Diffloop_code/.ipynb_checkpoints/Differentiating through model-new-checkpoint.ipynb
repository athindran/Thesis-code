```python
import numpy as np
import scipy
import torch
from matplotlib import pyplot as plt
import torch.optim as optim
from torch.autograd import Variable 

np.random.seed(0)
```

## Simple Pendulum (Experiment1)

```python
# Define the true system model
mtrue = 2
btrue = 1.5
ltrue = 0.8
g = 9.8
def func(x, t):
    u = 10*(np.sin(0.1*t))+10*np.sin(t)
    #noise = 0.0001*np.random.randn(1)
    noise = 0
    return [x[1], -(btrue/(mtrue*ltrue**2))*x[1]-(g/ltrue)*np.sin(x[0])+u/(mtrue*ltrue**2)+noise]
```

```python
# Excite the system to collect data
from scipy.integrate import odeint

t = np.linspace(0,100,10000)
x0 = [1,0]
sols = odeint(func, x0, t)
```

```python
# Create dataset of x,u,xdot
dataset = np.zeros((np.size(t),5))
for tind,sol in enumerate(sols):
    dataset[tind,0:2] = sol
    dataset[tind,2] = 10*(np.sin(0.1*t[tind]))+10*np.sin(t[tind])
    #dataset[tind,2] = 0
    dataset[tind,3:5] = np.array(func(sol,t[tind]))
```

```python
best = torch.tensor([3.0], requires_grad=True)
mest = torch.tensor([2.0], requires_grad=True)
lest = torch.tensor([4.0], requires_grad=True)

#optimizer1 = optim.SGD([best], lr=30, momentum=0.9)
#optimizer2 = optim.SGD([lest], lr=1, momentum=0.9)
#optimizer3 = optim.SGD([mest], lr=5, momentum=0.9)

#optimizer1.zero_grad()
#optimizer2.zero_grad()
#optimizer3.zero_grad()
#a1 = torch.tensor([2.0], requires_grad=True)
#a2 = torch.tensor([3.0], requires_grad=True)
#a3 = torch.tensor([1.5], requires_grad=True)
```

```python
niters = 10000
mvals = np.zeros((niters,))
lvals = np.zeros((niters,))
bvals = np.zeros((niters,))
lossvals = np.zeros((niters,))
lrm = 5e-3
lrb = 5e-3
lr = 1e-3
torchset = torch.tensor(dataset,requires_grad=False)  
for iters in range(niters):
  #if iters%5000==1:
  #  lrm=lrm/2
  #  lrb=lrb/2
  #  lr =lr/2
    
  mvals[iters] = mest.detach()
  lvals[iters] = lest.detach()
  bvals[iters] = best.detach()
  
  indices = np.random.randint(low=0, high=dataset.shape[0], size=64)
  #print(indices)
  #print(dataset[index,0],dataset[index,1],dataset[index,2],dataset[index,3],dataset[index,4])
  accest = -(best/(mest*lest**2))*torchset[indices,1]-(g/lest)*torch.sin(torchset[indices,0])+torchset[indices,2]/(mest*lest**2)
  #accest = -a1*torchset[indices,1]-a2*torch.sin(torchset[indices,0])+a3*torchset[indices,2]
  loss = ((accest-torchset[indices,4])**2).mean()

  loss.backward()
  #optimizer1.step()
  #optimizer2.step()
  #optimizer3.step()
  #optimizer1.zero_grad()
  #optimizer2.zero_grad()
  #optimizer3.zero_grad()
  with torch.no_grad():
    #a1 = torch.tensor([a1-lr*a1.grad],requires_grad=True)
    #a2 = torch.tensor([a2-lr*a2.grad],requires_grad=True)
    #a3 = torch.tensor([a3-lr*a3.grad],requires_grad=True)
    mest = torch.tensor([mest-lrm*mest.grad],requires_grad=True)
    best = torch.tensor([best-lrb*best.grad],requires_grad=True)
    lest = torch.tensor([lest-lr*lest.grad],requires_grad=True)
  lossvals[iters] = loss.detach()  
  
#print(a1)
#print(a2)
#print(a3)
print(mest)
print(best)
print(lest)
plt.figure()
plt.plot(lossvals)
plt.xlabel("Iteration")
plt.ylabel("Loss")

plt.figure()
plt.plot(mvals)
plt.plot(mtrue*np.ones(np.shape(mvals)))
plt.xlabel("Iteration")
plt.ylabel("Estimate of mass")

plt.figure()
plt.plot(lvals)
plt.plot(ltrue*np.ones(np.shape(lvals)))
plt.xlabel("Iteration")
plt.ylabel("Estimate of length")

plt.figure()
plt.plot(bvals)
plt.plot(btrue*np.ones(np.shape(bvals)))
plt.xlabel("Iteration")
plt.ylabel("Estimate of damping factor")
```

# Double Pendulum No damping (Experiment 1)

```python
m1 = 3
l1 = 1
m2 = 2
l2 = 0.5
g = 9.8
def func(x, t):
    u = 5*(np.sin(0.1*t))+5*np.sin(t)
    th1 = (-g*(2*m1+m2)*np.sin(x[0])-m2*g*np.sin(x[0]-2*x[1])-2*np.sin(x[0]-x[1])*m2*(x[3]**2*l2+x[2]**2*l1*np.cos(x[0]-x[1])))/(l1*(2*m1+m2-m2*np.cos(2*x[0]-2*x[1])))
    th2 = (2*np.sin(x[0]-x[1])*(x[2]**2*l1*(m1+m2)+g*(m1+m2)*np.cos(x[0])+x[3]**2*l2*m2*np.cos(x[0]-x[1]))+u)/(l1*(2*m1+m2-m2*np.cos(2*x[0]-2*x[1])))
    #u = 0
    return [x[2], x[3],th1,th2]
```

```python
# Excite the system to collect data
from scipy.integrate import odeint

t = np.linspace(0,100,10000)
x0 = [1,1,0,0]
sols = odeint(func, x0, t)
```

```python
# Create dataset of x,u,xdot
dataset = np.zeros((np.size(t),9))
for tind,sol in enumerate(sols):
    dataset[tind,0:4] = sol
    dataset[tind,4] = 5*(np.sin(0.1*t[tind]))+5*np.sin(t[tind])
    #dataset[tind,2] = 0
    dataset[tind,5:9] = np.array(func(sol,t[tind]))
```

```python
me1 = torch.tensor([5.0], requires_grad=True)
le1 = torch.tensor([2.0], requires_grad=True)
me2 = torch.tensor([1.0], requires_grad=True)
le2 = torch.tensor([1.0], requires_grad=True)
```

```python
niters = 2000
m1vals = np.zeros((niters,))
l1vals = np.zeros((niters,))
m2vals = np.zeros((niters,))
l2vals = np.zeros((niters,))
lossvals = np.zeros((niters,))
lrm = 1e-1
lr = 1e-2
torchset = torch.tensor(dataset,requires_grad=False)  
for iters in range(niters):
  if iters%1500==1:
    lrm=lrm/2
    lr =lr/2
    
  m1vals[iters] = me1.detach()
  l1vals[iters] = le1.detach()
  m2vals[iters] = me2.detach()
  l2vals[iters] = le2.detach()
  
  indices = np.random.randint(low=0, high=dataset.shape[0], size=64)

  the1 = (-g*(2*me1+me2)*np.sin(torchset[indices,0])-me2*g*torch.sin(torchset[indices,0]-2*torchset[indices,1])-2*torch.sin(torchset[indices,0]-torchset[indices,1])*me2*(torchset[indices,3]**2*l2+torchset[indices,2]**2*le1*torch.cos(torchset[indices,0]-torchset[indices,1])))/(le1*(2*me1+me2-me2*torch.cos(2*torchset[indices,0]-2*torchset[indices,1])))
  the2 = (2*torch.sin(torchset[indices,0]-torchset[indices,1])*(torchset[indices,2]**2*le1*(me1+me2)+g*(me1+me2)*torch.cos(torchset[indices,0])+torchset[indices,3]**2*le2*me2*torch.cos(torchset[indices,0]-torchset[indices,1]))+torchset[indices,4])/(le1*(2*me1+me2-me2*torch.cos(2*torchset[indices,0]-2*torchset[indices,1])))  
  #print(the1)
  loss1 = ((the1-torchset[indices,7])**2).mean()
  loss2 = ((the2-torchset[indices,8])**2).mean()
  loss = loss1+loss2
  #print(loss)
  loss.backward()
  with torch.no_grad():
    me1 = torch.tensor([me1-lrm*me1.grad],requires_grad=True)
    le1 = torch.tensor([le1-lr*le1.grad],requires_grad=True)
    me2 = torch.tensor([me2-lrm*me2.grad],requires_grad=True)
    le2 = torch.tensor([le2-lr*le2.grad],requires_grad=True)
  lossvals[iters] = loss.detach()  
  
print(me1)
print(le1)
print(me2)
print(le2)
plt.figure()
plt.plot(lossvals)
plt.xlabel("Iteration")
plt.ylabel("Loss")

plt.figure()
plt.plot(m1vals)
plt.plot(m1*np.ones(np.shape(m1vals)))
plt.xlabel("Iteration")
plt.ylabel("Estimate of mass 1")

plt.figure()
plt.plot(l1vals)
plt.plot(l1*np.ones(np.shape(l1vals)))
plt.xlabel("Iteration")
plt.ylabel("Estimate of length 1")

plt.figure()
plt.plot(m2vals)
plt.plot(m2*np.ones(np.shape(m2vals)))
plt.xlabel("Iteration")
plt.ylabel("Estimate of mass 2")

plt.figure()
plt.plot(l2vals)
plt.plot(l2*np.ones(np.shape(l2vals)))
plt.xlabel("Iteration")
plt.ylabel("Estimate of length 2")
```
