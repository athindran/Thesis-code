```python
import numpy as np
import scipy
import torch
from matplotlib import pyplot as plt
import torch.optim as optim
import torch.nn.functional as F
from torch.autograd import Variable 
import control
from control import matlab
from matplotlib import cm

np.random.seed(0)
if torch.cuda.is_available():  
  dev = "cuda:0" 
else:  
  dev = "cpu"  
device = torch.device(dev) 
```

```python
# Define the true system model
mtrue = 0.15
btrue = 0.1
ltrue = 0.5
#mtrue = 1
#btrue = 0
#ltrue = 1
g = 9.81

def pendulumt(x,u):
  noise = 0
  f = []
  for r in range(0,len(x)):
    f.append([x[r][1], -(btrue/(mtrue*ltrue**2))*x[r][1]+(g/ltrue)*torch.sin(x[r][0])])
  ft = torch.tensor(f,requires_grad=False)
  return ft

def pendulum(x,u):
  if u[0]>0.8:
    u[0]=0.8
  elif u[0]<-0.8:
    u[0]=-0.8
    
  return np.array([x[1], -(btrue/(mtrue*ltrue**2))*x[1]+(g/ltrue)*np.sin(x[0])+u[0]/(mtrue*ltrue**2)])
```

```python
A = np.array([[0,1],[(g/ltrue),-1*(btrue/(mtrue*ltrue**2))]])
B = np.array([[0],[1/(mtrue*ltrue**2)]])
C = np.eye(2)
D = 0

Q = np.diag([0.1,0.1])
R = 1
Dlqr = control.lqr(A,B,Q,R)
Klqr = np.array(Dlqr[0])
Slqr = np.array(Dlqr[1])
print(Klqr)
print(Slqr)
```

```python
nx = 100
ny = 100
x = np.linspace(-3.14, 3.14, nx)
y = np.linspace(-6, 6, ny)
xv, yv = np.meshgrid(x, y)
xvnext = np.array(xv)
yvnext = np.array(yv)
verifies = np.zeros((nx,ny),dtype=np.bool)
V = np.zeros((nx,ny))
Vdot = np.zeros((nx,ny))

#plt.figure()
for i in range(nx):
    for j in range(ny):
      xvec = np.array([[xv[i,j]],[yv[i,j]]])
      xvecdot = np.array(pendulum(xvec,-np.dot(Klqr,xvec)))
      xvnext[i,j] = xvecdot[0]
      yvnext[i,j] = xvecdot[1]
      V[i,j] = np.dot(np.dot(xvec.T,Slqr),xvec)
      Vdot[i,j] = np.dot(np.dot(xvecdot.T,Slqr),xvec)  
      if V[i,j]>=0 and Vdot[i,j]<0:
        verifies[i,j] = True
fig = plt.figure()
plt.scatter(xv[verifies],yv[verifies],color='w')
plt.scatter(xv[np.logical_not(verifies)],yv[np.logical_not(verifies)],color='k')
contours = plt.contour(xv,yv,V,[0.3,1,5,10,20,100])
plt.clabel(contours, inline=1, fontsize=10)
plt.streamplot(xv,yv,xvnext,yvnext, color=('gray'), linewidth=0.5,
                  density=0.5, arrowstyle='-|>', arrowsize=1.5)
plt.xlabel('Angle (rad)')
plt.ylabel('Angular velocity (rad/s)')
plt.show()
```

```python
T = 1000
sols = np.zeros((T,3))
x = np.array([2,5]);


for t in np.arange(0,T):
    sols[t,0:2] = x;
    u = -np.dot(Klqr,x)
    #print(u)
    xdot = pendulum(x,u)
    
    lyaph = np.dot(Slqr,x)
    sols[t,2] = np.dot(x,lyaph.ravel())
    x = x+0.01*xdot;
    
plt.figure()
plt.plot(0.01*np.arange(0,T),sols[:,0])
plt.title("LQR State 0")

plt.figure()
plt.plot(0.01*np.arange(0,T),sols[:,1])
plt.title("LQR State 1")
print(sols[-1,0])
plt.figure()
plt.plot(0.01*np.arange(0,T),sols[:,2])
plt.title("lyapunov function")
print(sols[-1,2])
```

```python
class lyapNetquad(torch.nn.Module):
  def __init__(self):
    super(lyapNetquad, self).__init__()
    torch.manual_seed(2)
    self.lyaphalf1 = torch.nn.Linear(2,7,2,bias=False)
    self.ls = torch.randn(1,1) 
  
  def forward(self,x):
    lyaph = self.lyaphalf1(x)
    return lyaph,self.ls

class controllerLinear(torch.nn.Module):
   def __init__(self):
    super(controllerLinear, self).__init__()
    torch.manual_seed(2)
    self.control = torch.nn.Linear(2,1,bias=False)
    self.control.weight = torch.nn.Parameter(torch.tensor(([[-1.53,-0.46]])))
  
   def forward(self,x):
    contr = self.control(x)
    return contr  

class controllermed(torch.nn.Module):
   def __init__(self):
    super(controllermed, self).__init__()
    torch.manual_seed(2)
    self.control1 = torch.nn.Linear(2,5,bias=False)
    self.control2 = torch.nn.Linear(5,1,bias=False)
  
   def forward(self,x):
    sigmoid = torch.nn.Tanh()
    contr1 = sigmoid(self.control1(x))
    contr = self.control2(contr1)
    return contr  
        
class controllerlar(torch.nn.Module):
   def __init__(self):
    super(controllerlar, self).__init__()
    torch.manual_seed(2)
    self.control1 = torch.nn.Linear(2,5,bias=False)
    self.control2 = torch.nn.Linear(5,7,bias=False)
    self.control3 = torch.nn.Linear(7,1,bias=False)
  
   def forward(self,x):
    sigmoid = torch.nn.Tanh()  
    contr1 = sigmoid(self.control1(x))
    contr2 = sigmoid(self.control2(contr1))
    contr = self.control3(contr2)
    return contr         

class lyapNetmed(torch.nn.Module):
  def __init__(self):
    super(lyapNetmed, self).__init__()
    torch.manual_seed(2)
    self.lyaphalf1 = torch.nn.Linear(2,5,bias=False)
    self.lyaphalf2 = torch.nn.Linear(5,10,bias=False)
    #self.ls = torch.tensor([1000.0], requires_grad=True)
  
  def forward(self,x):
    sigmoid = torch.nn.Tanh()    
    lyaph1 = sigmoid(self.lyaphalf1(x))
    lyaph = self.lyaphalf2(lyaph1)
    return lyaph

class lyapNetlarge(torch.nn.Module):
  def __init__(self):
    super(lyapNetlarge, self).__init__()
    torch.manual_seed(2)
    self.lyaphalf1 = torch.nn.Linear(2,32,bias=False)
    self.lyaphalf2 = torch.nn.Linear(32,64,bias=False)
    self.lyaphalf3 = torch.nn.Linear(64,10,bias=False)
  
  def forward(self,x):
    sigmoid = torch.nn.Tanh()    
    lyaph1 = sigmoid(self.lyaphalf1(x))
    lyaph2 = sigmoid(self.lyaphalf2(lyaph1))
    lyaph = self.lyaphalf3(lyaph2)
    return lyaph
```

```python
    ## takes in a module and applies the specified weight initialization
    def weights_init_normal(m):
        '''Takes in a module and initializes all linear layers with weight
           values taken from a normal distribution.'''

        classname = m.__class__.__name__
        # for every Linear layer in a model
        if classname.find('Linear') != -1:
            y = m.in_features
        # m.weight.data shoud be taken from a normal distribution
            m.weight.data.normal_(0.0,1)
```

```python
def lyaptraining(dataset,Kcontrol,ls,lyaph_net,max_iters,Q,R,batch_size,learning_rate):
  iters = 0
  costs = []
  torch.manual_seed(2)
  
  optimizer = torch.optim.Adam((lyaph_net.parameters()), lr=learning_rate, weight_decay=0.7)
  roalearn = torch.optim.Adam([ls],lr=learning_rate)
  while iters<max_iters:
    indices = np.random.randint(low=0, high=dataset.shape[0], size=batch_size)
    x = dataset[indices,:].clone().detach().requires_grad_(False)
    x.to(device)
    lyaph = lyaph_net(x.cuda())
    #print(x.shape)
    u = Kcontrol(x.cuda())
    u = torch.clamp(u,-0.8,0.8)
    xdot = pendulumt(x,u)
    xdotu = torch.zeros((batch_size,2))
    xdotu[:,1] = u[:,0]/(mtrue*ltrue**2)
    xdot = xdot+xdotu;
    xnext = x+0.01*xdot;
    xnext.to(device)
    lyaphnext = lyaph_net(xnext.cuda())
    lyaph = torch.reshape(lyaph,(100,2,5))
    xs = torch.unsqueeze(x.cuda(),1)
    lyapfunch = torch.squeeze(torch.matmul(xs,lyaph),1)
    lyapfunc = torch.sum(lyapfunch**2,dim=1)+0.1*torch.sum(x.cuda()**2,dim=1)
    
    xsnext = torch.unsqueeze(xnext.cuda(),1)
    lyaphnext = torch.reshape(lyaphnext,(100,2,5))
    lyapfuncnh = torch.squeeze(torch.matmul(xsnext,lyaphnext),1)
    lyapfuncnext = torch.sum(lyapfuncnh**2,dim=1)+0.1*torch.sum(xnext.cuda()**2,dim=1)
    lyapder = lyapfuncnext-lyapfunc
    lqrcost = torch.sum(torch.mul(torch.mm(x.cuda(),Q.cuda()),x.cuda()),dim=1)+torch.sum(R.cuda()*torch.mul(u,u),dim=1)
    
    cost =  (torch.div(0.5*(torch.sign(ls.cuda()-lyapfunc)+1)*(F.relu(lyapder+0.9*lqrcost+1)),0.001*(lyapfunc+0.1))).mean()\
            +0.1*(torch.sign(lyapder)*(ls.cuda()-lyapfunc)).mean()
  
    if iters%1000==0:
      print(iters,"Loss=",cost.item(),"Lyapunovfunc",lyapfunc.mean().data,"Lyapunov deriv",lyapder.mean().data,"LQR cost",lqrcost.mean().data,"ls",ls)
    costs.append(cost.item())
    optimizer.zero_grad()
    roalearn.zero_grad()
    cost.backward()
    optimizer.step()
    roalearn.step()
    iters = iters+1
  return costs,Kcontrol,lyaph_net,ls

def controltraining(dataset,Kcontrol,ls,lyaph_net,max_iters,Q,R,batch_size,learning_rate):
  iters = 0
  costs = []
  torch.manual_seed(2)
  
  optimizer = torch.optim.Adam((Kcontrol.parameters()), lr=learning_rate, weight_decay=0.3)
  roalearn = torch.optim.Adam([ls],lr=learning_rate)

  while iters<max_iters:
    indices = np.random.randint(low=0, high=dataset.shape[0], size=batch_size)
    x = dataset[indices,:].clone().detach().requires_grad_(False)
    x.to(device)
    lyaph = lyaph_net(x.cuda())
    
    u = Kcontrol(x.cuda())
    u = torch.clamp(u,-0.8,0.8)
    xdot = pendulumt(x,u)
    xdotu = torch.zeros((batch_size,2))
    xdotu[:,1] = u[:,0]/(mtrue*ltrue**2)
    xdot = xdot+xdotu;
    xnext = x+0.01*xdot;
    xnext.to(device)
    lyaphnext = lyaph_net(xnext.cuda())
    
    lyaph = torch.reshape(lyaph,(100,2,5))
    xs = torch.unsqueeze(x.cuda(),1)
    lyapfunch = torch.squeeze(torch.matmul(xs,lyaph),1)
    lyapfunc = torch.sum(lyapfunch**2,dim=1)+0.1*torch.sum(x.cuda()**2,dim=1)
    
    xsnext = torch.unsqueeze(xnext.cuda(),1)
    lyaphnext = torch.reshape(lyaphnext,(100,2,5))
    lyapfuncnh = torch.squeeze(torch.matmul(xsnext,lyaphnext),1)
    lyapfuncnext = torch.sum(lyapfuncnh**2,dim=1)+0.1*torch.sum(xnext.cuda()**2,dim=1)
    lyapder = lyapfuncnext-lyapfunc
    lqrcost = torch.sum(torch.mul(torch.mm(x.cuda(),Q.cuda()),x.cuda()),dim=1)+torch.sum(R.cuda()*torch.mul(u,u),dim=1)
    
    cost =  (torch.div(0.5*(torch.sign(ls.cuda()-lyapfunc)+1)*(F.relu(lyapder+0.9*lqrcost+1)),0.001*(lyapfunc+0.1))).mean()\
            +0.1*(torch.sign(lyapder)*(ls.cuda()-lyapfunc)).mean()

    if iters%1000==0:
      print(iters,"Loss=",cost.item(),"Lyapunovfunc",lyapfunc.mean().data,"Lyapunov deriv",lyapder.mean().data,"LQR cost",lqrcost.mean().data,"ls",ls)
    costs.append(cost.item())
    optimizer.zero_grad()
    roalearn.zero_grad()
    cost.backward()
    optimizer.step()
    roalearn.step()
    iters = iters+1
  return costs,Kcontrol,lyaph_net,ls
```
