```python
import numpy as np
import scipy
import torch
from matplotlib import pyplot as plt
import torch.optim as optim
import torch.nn.functional as F
from torch.autograd import Variable 
import control
from control import matlab
from matplotlib import cm

np.random.seed(0)
dev = "cpu"  
device = torch.device(dev) 
```

```python
sysA = 1.010
#sysA = 0.99
sysB = 0.125
sysC = 0.1608
sysD = 0

kp = 4.0
ki = 16.0
```

```python
x = 0

ref = np.ones((1500,))
ref[200:400] = 2;
ref[400:600] = 0;
ref[600:1000] = -2.1;
ref[1000:1400] = 2.0;
ref[1400:1500] = 3.0;

ysat = np.zeros((1500,))
uvalssat = np.zeros((1500,))
xcont = 0;
u = 0;
for i in range(1500):
   ysat[i] = sysC*x+sysD*u;
   xcont = xcont+(ref[i]-ysat[i]);
   u = 0.01*ki*xcont+kp*(ref[i]-ysat[i]);
   if(u>4):
     u=4;
   elif(u<-4):
     u=-4;  
   uvalssat[i] = u;
   x = sysA*x+sysB*u;

ynosat = np.zeros((1500,))
uvalsnosat = np.zeros((1500,))
xcont = 0;
u = 0;
for i in range(1500):
   ynosat[i] = sysC*x+sysD*u;
   xcont = xcont+(ref[i]-ynosat[i]);
   u = 0.01*ki*xcont+kp*(ref[i]-ynosat[i]); 
   uvalsnosat[i] = u;
   x = sysA*x+sysB*u;
```

```python
# Use policy gradient to tune kp,ki
ysat = np.zeros((1500,))
uvalssat = np.zeros((1500,))
uvalspol = np.zeros((1500,))

kppg = kp
kipg = ki
nupdates = 10
nepisodes = 128
updatecosts = np.zeros((nupdates,))
for update in range(nupdates):
  costs = np.zeros((nepisodes,))
  kpder = np.zeros((nepisodes,))
  kider = np.zeros((nepisodes,))  
  for episode in range(nepisodes):
    ref = np.ones((1500,))  
    ref[200:400] = 2;
    ref[400:600] = 0;
    ref[600:1000] = -2.1;
    ref[1000:1400] = 2.0; 
    ref[1400:1500] = 3.0;
   
    xcont = 0;
    u = 0;
    x = 0;
    sigma = 0.5
    for t in range(1500):
      ysat[t] = sysC*x+sysD*u;
      xcont = xcont+(ref[t]-ysat[t]);
      uvalspol[t] = 0.01*kipg*xcont+kppg*(ref[t]-ysat[t])  
      u = uvalspol[t]+sigma*np.random.randn()
      uvalssat[t] = u
    
      if(u>4):
        u=4
      elif(u<-4):
        u=-4
      x = sysA*x+sysB*u;                                
      costs[episode] += (ref[t]-ysat[t])**2
      kpder[episode] += (ref[t]-ysat[t])*(uvalssat[t]-uvalspol[t])/(2*sigma)
      kider[episode] += 0.01*xcont*(uvalssat[t]-uvalspol[t])/(2*sigma)
      #kpder[episode] += -(kpt-kppg)*(ref[i]-ysat[i])/(2*sigma)
      #kider[episode] += -(kit-kipg)*0.01*xcont/(2*sigma)
    #print("Episode cost: ",costs[episode])    
  # Update kp,ki
  lr = 0.1
  updatecosts[update] = np.mean(costs)
  print("Update cost",updatecosts[update])
  kppg = kppg-lr*np.clip(np.mean(kpder*(costs-np.mean(costs))),-1000,1000)
  kipg = kipg-lr*np.clip(np.mean(kider*(costs-np.mean(costs))),-1000,1000)
  print(kppg)
  print(kipg)  
  #if(kppg<0):
  #  kppg = 0
  #if(kipg<0):
  #  kipg = 0                                 
                                     
```

```python
# Use policy gradient to tune kp,ki
kppg = kp
kipg = ki
satpg = 6.0
nupdates = 50
nepisodes = 128
updatecosts = np.zeros((nupdates,))

for update in range(nupdates):
  costs = np.zeros((nepisodes,))
  kpder = np.zeros((nepisodes,))
  kider = np.zeros((nepisodes,))
  satder = np.zeros((nepisodes,)) 
  for episode in range(nepisodes):
    ref = refs[5+episode%5,:]
    xcont = 0;
    u = 0;
    x = 0;
    sigma = 1
    noise = 0.1*np.random.randn(1000,1)
    ypg = np.zeros((1000,1))
    uvalspg = np.zeros((1000,1))
    uvalspol = np.zeros((1000,1))
    uvalsfinal = np.zeros((1000,1))
    for t in range(1000):
      ypg[t] = sysC*x+sysD*u+noise[t]
      xcont = xcont+(ref[t]-ypg[t]);
      ugain = 0.01*kipg*xcont+kppg*(ref[t]-ypg[t])
      upresat = np.tanh(ugain)
      uvalspol[t] = satpg*upresat
      u = uvalspol[t]+sigma*np.random.randn()
      uvalspg[t] = u
    
      if(u>3):
        u=3
      elif(u<-3):
        u=-3
      uvalsfinal[t] = u  
      x = sysA*x+sysB*u;                                
      costs[episode] += min((ref[t]-ypg[t])**2,500)
      kpder[episode] += (ref[t]-ypg[t])*(1-upresat**2)*(uvalspg[t]-uvalspol[t])/(2*sigma)
      kider[episode] += 0.01*xcont*(1-upresat**2)*(uvalspg[t]-uvalspol[t])/(2*sigma)
      satder[episode] += (upresat-(1-upresat**2)*ugain/satpg)*(uvalspg[t]-uvalspol[t])/(2*sigma)
      #kpder[episode] += -(kpt-kppg)*(ref[i]-ysat[i])/(2*sigma)
      #kider[episode] += -(kit-kipg)*0.01*xcont/(2*sigma)
    #print("Episode cost: ",costs[episode])    
  # Update kp,ki
  lr = 0.001
  updatecosts[update] = np.mean(costs)
  print("Update cost",updatecosts[update]) 
  kppg = kppg-lr*np.clip(np.mean(kpder*(costs-np.mean(costs))),-10000,10000)
  kipg = kipg-lr*np.clip(np.mean(kider*(costs-np.mean(costs))),-10000,10000)
  satpg = satpg-lr*np.clip(np.mean(satder*(costs-np.mean(costs))),-10000,10000)
  print(kppg,kipg,satpg)
  #kppg = np.clip(kppg,0,200)
  #kipg = np.clip(kipg,0,200)
```
