```python
import numpy as np
import scipy
import torch
from matplotlib import pyplot as plt
import torch.optim as optim
import torch.nn.functional as F
from torch.autograd import Variable 
import control
from control import matlab
from matplotlib import cm

np.random.seed(0)
if torch.cuda.is_available():  
  dev = "cuda:0" 
else:  
  dev = "cpu"  
device = torch.device(dev) 
```

```python
M = 1
m = 0.1
b = 0
L = 0.5
I = 0.006
g = 9.81

def pendulumcartt(x,u):
  noise = 0
  f = []
  xdot = torch.zeros(x.shape,requires_grad=False, device=dev)  
  for r in range(0,len(x)):
    xdot[r][0] = x[r][2]
    xdot[r][1] = x[r][3]
    xdot[r][3] = ((M+m)*m*g*L*torch.sin(x[r][1])+m*L*(-m*L*(x[r][3]**2)*torch.sin(x[r][1])-b*x[r][2])*torch.cos(x[r][1]))/((M+m)*(I+m*L*L)-m*m*L*L*torch.cos(x[r][1])**2) 
    xdot[r][2] = (-m*L*x[r][3]*x[r][3]*torch.sin(x[r][1])+m*L*xdot[r][3]*torch.cos(x[r][1])-b*x[r][2])/(M+m)
  noise = 0.1*torch.randn(x.shape)
  noise.to(device)
  return xdot+noise.cuda()

# x - x, theta, xdot, thetadot
def pendulumcart(x,u):
  if u[0]>15:
    u[0]=15
  elif u[0]<-15:
    u[0]=-15
  xdot = np.zeros((4,))
  xdot[0] = x[2]
  xdot[1] = x[3]
  xdot[3] = ((M+m)*m*g*L*np.sin(x[1])+m*L*(u[0]-m*L*(x[3]**2)*np.sin(x[1])-b*x[2])*np.cos(x[1]))/((M+m)*(I+m*L*L)-m*m*L*L*np.cos(x[1])**2) 
  #print(u[0]/(M+m))
  xdot[2] = (u[0]-m*L*x[3]*x[3]*np.sin(x[1])+m*L*xdot[3]*np.cos(x[1])-b*x[2])/(M+m)
  #xdot[2] = xdot[2]+u[0]/(M+m)  
  return xdot+0.1*np.random.randn(4,)
```

```python
A = np.array([[0,0,1,0],[0,0,0,1],[0,(m*m*g*L*L)/(I*(M+m)+M*m*L*L),-((I+m*L*L)*b)/(I*(M+m)+M*m*L*L),0],[0,(m*g*L*(M+m))/(I*(M+m)+M*m*L*L),(-m*L*b)/(I*(M+m)+M*m*L*L),0]])
B = np.array([[0],[0],[(I+m*L*L)/(I*(M+m)+M*m*L*L)],[(m*L)/(I*(M+m)+M*m*L*L)]])
C = np.eye(4)
D = 0
print(A)
Q = np.diag([1,1,1,1])
R = 1
Dlqr = control.lqr(A,B,Q,R)
Klqr = np.array(Dlqr[0])
Slqr = np.array(Dlqr[1])
print(Klqr)
print(Slqr)
#[ 2.0007045 -1.5074435  0.6653867 -0.8102326]
# [-1.5074435 16.63473   -2.0351412  7.589846 ]
# [ 0.6653867 -2.0351412  3.1598542 -3.5465822]
# [-0.8102326  7.589846  -3.5465822  5.871508 ]
```

```python
class lyapNetquad(torch.nn.Module):
  def __init__(self):
    super(lyapNetquad, self).__init__()
    torch.manual_seed(2)
    self.lyaphalf = torch.nn.Linear(4,4,bias=False)
  
  def forward(self,x):
    #lyaph = self.lyaphalf(x)
    return self.lyaphalf.weight.T

class controllerLinear(torch.nn.Module):
   def __init__(self):
    super(controllerLinear, self).__init__()
    torch.manual_seed(2)
    self.control = torch.nn.Linear(4,1,bias=False)
    self.control.weight = torch.nn.Parameter(torch.tensor([[1,-31.46,2.28,-7.77]]))
  
   def forward(self,x):
    contr = self.control(x)
    return contr

class controllermed(torch.nn.Module):
   def __init__(self):
    super(controllermed, self).__init__()
    torch.manual_seed(2)
    self.control1 = torch.nn.Linear(4,5,bias=False)
    self.control2 = torch.nn.Linear(5,1,bias=False)
  
   def forward(self,x):
    sigmoid = torch.nn.Tanh()
    contr1 = sigmoid(self.control1(x))
    contr = self.control2(contr1)
    return contr

class lyapNetmed(torch.nn.Module):
  def __init__(self):
    super(lyapNetmed, self).__init__()
    torch.manual_seed(2)
    self.lyaphalf1 = torch.nn.Linear(4,5,bias=False)
    self.lyaphalf2 = torch.nn.Linear(5,20,bias=False)
    #self.ls = torch.tensor([1000.0], requires_grad=True)
  
  def forward(self,x):
    sigmoid = torch.nn.Tanh()    
    lyaph1 = sigmoid(self.lyaphalf1(x))
    lyaph = self.lyaphalf2(lyaph1)
    return lyaph

class lyapNetlarge(torch.nn.Module):
  def __init__(self):
    super(lyapNetlarge, self).__init__()
    torch.manual_seed(2)
    self.lyaphalf1 = torch.nn.Linear(4,32,bias=False)
    self.lyaphalf2 = torch.nn.Linear(32,48,bias=False)
    self.lyaphalf3 = torch.nn.Linear(48,56,bias=False)
    self.lyaphalf4 = torch.nn.Linear(56,64,bias=False)
    self.lyaphalf5 = torch.nn.Linear(64,20,bias=False)
  
  def forward(self,x):
    sigmoid = torch.nn.Tanh()    
    lyaph1 = sigmoid(self.lyaphalf1(x))
    lyaph2 = sigmoid(self.lyaphalf2(lyaph1))
    lyaph3 = sigmoid(self.lyaphalf3(lyaph2))
    lyaph4 = sigmoid(self.lyaphalf4(lyaph3))
    lyaph = self.lyaphalf5(lyaph4)
    return lyaph
```

```python
def lyaptraining(dataset,Kcontrol,ls,lyaph_net,max_iters,Q,R,batch_size,learning_rate,controllertrain):
  iters = 0
  costs = []
  torch.manual_seed(2)
  
  if not controllertrain:
    optimizer = torch.optim.Adam((lyaph_net.parameters()), lr=learning_rate, weight_decay=0.01)
  else:
    optimizer = torch.optim.Adam((Kcontrol.parameters()), lr=learning_rate)

  roalearn = torch.optim.Adam([ls],lr=learning_rate)
  while iters<max_iters:
    indices = np.random.randint(low=0, high=dataset.shape[0], size=batch_size)
    x = dataset[indices,:].clone().detach().requires_grad_(False)
    #print(dataset[indices+1,:])
    x.to(device)
    xc = x.cuda()
    lyaph = lyaph_net(xc)
    #print(lyaph.shape)
    #print(x.shape)
    u = Kcontrol(xc)
    lsn = F.relu(ls.cuda())
    u = torch.clamp(u,-15,15)
    uc = u.cpu()
    xdot = pendulumcartt(xc,0)
    xdotu = torch.zeros((batch_size,4),requires_grad=True)
    xdotu.to(device)
    xdotuc = xdotu.cuda()
    xdotuc[:,3] = torch.div((m*L*u[:,0]*torch.cos(xc[:,1])),((M+m)*(I+m*L*L)-m*m*L*L*torch.cos(xc[:,1])**2))
    xdotuc[:,2] = (m*L*xdotuc[:,3]*torch.cos(xc[:,1]))/(M+m)
    xdotuc[:,2] = xdotuc[:,2]+u[:,0]/(M+m)
    #print(xdot)
    #print(xdotu)
    xdot = xdot+xdotuc
    
    #print(xdot)
    #xs = xc.cpu().detach().numpy()
    #us = u.cpu().detach().numpy()
    #xdoto = pendulumcart(xs[0,:],us[0,:])
    #print(xdoto)
    
    xnext = x.cuda()+0.01*xdot.cuda();
    #print(xnext)
    #torch.clamp(xnext[:,1],-8,8)
    xnext.to(device)
    #print(x ,xdot, xnext)
    lyaphnext = lyaph_net(xnext.cuda())
    lyaph = torch.reshape(lyaph,(120,4,5))
    xs = torch.unsqueeze(x.cuda(),1)
    lyapfunch = torch.squeeze(torch.matmul(xs,lyaph),1)
    #lyapfunch = torch.matmul(x.cuda(),lyaph)
    lyapfunc = torch.sum(lyapfunch**2,dim=1)+0.1*torch.sum(x.cuda()**2,dim=1)
    
    xsnext = torch.unsqueeze(xnext.cuda(),1)
    lyaphnext = torch.reshape(lyaphnext,(120,4,5))
    lyapfuncnh = torch.squeeze(torch.matmul(xsnext,lyaphnext),1)
    #lyapfuncnh = torch.matmul(xnext.cuda(),lyaph)
    lyapfuncnext = torch.sum(lyapfuncnh**2,dim=1)+0.1*torch.sum(xnext.cuda()**2,dim=1)
    lyapder = lyapfuncnext-lyapfunc
    lqrcost = torch.sum(torch.mul(torch.mm(x.cuda(),Q.cuda()),x.cuda()),dim=1)+torch.sum(R.cuda()*torch.mul(u,u),dim=1)
    
    #cost =  (torch.div(0.5*(torch.sign(lsn.cuda()-lyapfunc)+1)*(F.relu(lyapder+0.9*lqrcost+1)),0.0001*(lyapfunc+10))).mean()\
            #+0.1*(torch.sign(lyapder)*(lsn.cuda()-lyapfunc)).mean()
    cost = 10000*(F.relu(lyapder+0.9*lqrcost+1)).mean()
    if iters%1000==0:
      print(iters,"Loss=",cost.item(),"Lyapunovfunc",lyapfunc.mean().data,"Lyapunov deriv",lyapder.mean().data,"LQR cost",lqrcost.mean().data,"ls",lsn)
    costs.append(cost.item())
    optimizer.zero_grad()
    roalearn.zero_grad()
    cost.backward()
    optimizer.step()
    roalearn.step()
    iters = iters+1
  return costs,Kcontrol,lyaph_net,ls
```

```python
def controltraining(dataset,Kcontrol,ls,lyaph_net,max_iters,Q,R,batch_size,learning_rate,sat,controllertrain):
  iters = 0
  costs = []
  torch.manual_seed(2)
  
  if not controllertrain:
    optimizer = torch.optim.Adam((lyaph_net.parameters()), lr=learning_rate, weight_decay=0)
  else:
    optimizer = torch.optim.Adam((Kcontrol.parameters()), lr=learning_rate)
    soptimizer = torch.optim.Adam([sat], lr=learning_rate)

  roalearn = torch.optim.Adam([ls],lr=learning_rate)
  while iters<max_iters:
    indices = np.random.randint(low=0, high=dataset.shape[0], size=batch_size)
    x = dataset[indices,:].clone().detach().requires_grad_(False)
    #print(dataset[indices+1,:])
    x.to(device)
    xc = x.cuda()
    lyaph = lyaph_net(xc)
    #print(lyaph.shape)
    #print(x.shape)
    satc = sat.cuda()
    u = satc*torch.tanh(Kcontrol(xc)/satc)
    lsn = F.relu(ls.cuda())
    u = torch.clamp(u,-15,15)
    xdot = pendulumcartt(xc,0)
    xdotu = torch.zeros((batch_size,4),requires_grad=True)
    xdotu.to(device)
    xdotuc = xdotu.cuda()
    xdotuc[:,3] = torch.div((m*L*u[:,0]*torch.cos(xc[:,1])),((M+m)*(I+m*L*L)-m*m*L*L*torch.cos(xc[:,1])**2))
    xdotuc[:,2] = (m*L*xdotuc[:,3]*torch.cos(xc[:,1]))/(M+m)
    xdotuc[:,2] = xdotuc[:,2]+u[:,0]/(M+m)
    #print(xdot)
    #print(xdotu)
    xdot = xdot+xdotuc
    
    #print(xdot)
    #xs = xc.cpu().detach().numpy()
    #us = u.cpu().detach().numpy()
    #xdoto = pendulumcart(xs[0,:],us[0,:])
    #print(xdoto)
    
    xnext = x.cuda()+0.01*xdot.cuda();
    #print(xnext)
    #torch.clamp(xnext[:,1],-8,8)
    xnext.to(device)
    #print(x ,xdot, xnext)
    lyaphnext = lyaph_net(xnext.cuda())
    lyaph = torch.reshape(lyaph,(120,4,5))
    xs = torch.unsqueeze(x.cuda(),1)
    lyapfunch = torch.squeeze(torch.matmul(xs,lyaph),1)
    #lyapfunch = torch.matmul(x.cuda(),lyaph)
    lyapfunc = torch.sum(lyapfunch**2,dim=1)+0.1*torch.sum(x.cuda()**2,dim=1)
    
    xsnext = torch.unsqueeze(xnext.cuda(),1)
    lyaphnext = torch.reshape(lyaphnext,(120,4,5))
    lyapfuncnh = torch.squeeze(torch.matmul(xsnext,lyaphnext),1)
    #lyapfuncnh = torch.matmul(xnext.cuda(),lyaph)
    lyapfuncnext = torch.sum(lyapfuncnh**2,dim=1)+0.1*torch.sum(xnext.cuda()**2,dim=1)
    lyapder = lyapfuncnext-lyapfunc
    lqrcost = torch.sum(torch.mul(torch.mm(x.cuda(),Q.cuda()),x.cuda()),dim=1)+torch.sum(R.cuda()*torch.mul(u,u),dim=1)
    
    #cost =  (torch.div(0.5*(torch.sign(lsn.cuda()-lyapfunc)+1)*(F.relu(lyapder+0.9*lqrcost+1)),0.0001*(lyapfunc+10))).mean()\
    #        #+0.1*(torch.sign(lyapder)*(lsn.cuda()-lyapfunc)).mean()
    #cost = (0.5*(torch.sign(lsn.cuda()-lyapfunc)+1)*F.relu(lyapder+0.9*lqrcost+1)).mean()
    cost = 10000*(F.relu(lyapder+0.9*lqrcost+1)).mean()
    if iters%1000==0:
      print(iters,"Loss=",cost.item(),"Lyapunovfunc",lyapfunc.mean().data,"Lyapunov deriv",lyapder.mean().data,"LQR cost",lqrcost.mean().data,"ls",lsn,"sat",sat)
      #print(Kcontrol.control.weight)
    costs.append(cost.item())
    optimizer.zero_grad()
    roalearn.zero_grad()
    soptimizer.zero_grad()
    cost.backward()
    optimizer.step()
    roalearn.step()
    soptimizer.step()
    iters = iters+1
  return costs,Kcontrol,lyaph_net,ls, sat
```

```python
def generatedataset(startstates):
    dataset = []
    #print(startstates.shape[0])
    for i in range(startstates.shape[0]):
      #print(i)
      T = 2000
      x = startstates[i,:];
      startdataset = []
      for t in np.arange(0,T):
        xdot = pendulumcart(x,-np.dot(Klqr,x))
        x = x+0.01*xdot;
        #if(t%3==0):
        startdataset.append(list(x))            
        #print(x)
        if(np.linalg.norm(x)<0.1):
          break;
      if(np.linalg.norm(x)<0.1):
         dataset = dataset + startdataset
    return dataset  
```

```python
nx = 20
ny = 20
nz = 20
nu = 20

x = np.linspace(-2.0, 2.0, nx)
y = np.linspace(-0.85, 0.85, ny)
z = np.linspace(-2.0,2.0,nz)
u = np.linspace(-0.85,0.85,nu)
xv, yv, zv, uv = np.meshgrid(x, y, z, u)
xvnext = np.array(xv)
yvnext = np.array(yv)
zvnext = np.array(zv)
uvnext = np.array(uv)
verifies = np.zeros((nx,ny,nz,nu),dtype=np.bool)
V = np.zeros((nx,ny,nz,nu))
Vdot = np.zeros((nx,ny,nz,nu))

#plt.figure()
for i in range(nx):
    for j in range(ny):
      for k in range(nz):
        for l in range(nu):
          xvec = np.array([[xv[i,j,k,l]],[yv[i,j,k,l]],[zv[i,j,k,l]],[uv[i,j,k,l]]])
          xt = torch.tensor([xv[i,j,k,l],yv[i,j,k,l],zv[i,j,k,l],uv[i,j,k,l]]).float()
          xt.to(device)
          u = sat.cuda()*torch.tanh(Kcontrol2(xt.cuda())/sat.cuda())
          xvecdot = np.array(pendulumcart(xvec,[u.cpu().detach().numpy()]))
          xvnext[i,j,k,l] = xvecdot[0]
          yvnext[i,j,k,l] = xvecdot[1]
          zvnext[i,j,k,l] = xvecdot[2]
          uvnext[i,j,k,l] = xvecdot[3]
        
          lyah = lyaph_net(xt.cuda())
          lyah = torch.reshape(lyah,(4,5))
          lya = torch.sum(torch.matmul(xt.cuda(),lyah)**2)+0.1*torch.sum(xt.cuda()**2)
    
          xt = torch.tensor([xv[i,j,k,l]+0.01*xvnext[i,j,k,l],yv[i,j,k,l]+0.01*yvnext[i,j,k,l],zv[i,j,k,l]+0.01*zvnext[i,j,k,l],uv[i,j,k,l]+0.01*uvnext[i,j,k,l]]).float()
          xt.to(device)
          lyah = lyaph_net(xt.cuda())
          lyah = torch.reshape(lyah,(4,5))
          lyan = torch.sum(torch.matmul(xt.cuda(),lyah)**2)+0.1*torch.sum(xt.cuda()**2)
      
          V[i,j,k,l] = lya.cpu().detach().numpy()
          Vnext = lyan.cpu().detach().numpy()
          Vdot[i,j,k,l] = Vnext-V[i,j,k,l]
          if V[i,j,k,l]>=0 and Vdot[i,j,k,l]<0:
            verifies[i,j,k,l] = True
```
