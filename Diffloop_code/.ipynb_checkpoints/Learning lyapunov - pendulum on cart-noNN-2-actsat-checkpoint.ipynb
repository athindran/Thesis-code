```python
import numpy as np
import scipy
import torch
from matplotlib import pyplot as plt
import torch.optim as optim
import torch.nn.functional as F
from torch.autograd import Variable 
import control
from control import matlab
from matplotlib import cm

np.random.seed(0)
if torch.cuda.is_available():  
  dev = "cuda:0" 
else:  
  dev = "cpu"  
device = torch.device(dev) 
```

```python
M = 1
m = 0.1
b = 0
L = 0.5
I = 0.006
g = 9.81

def pendulumcartt(x,u):
  noise = 0
  f = []
  xdot = 0.2*torch.randn(x.shape,requires_grad=False, device=dev)  
  for r in range(0,len(x)):
    xdot[r][0] = x[r][2]
    xdot[r][1] = x[r][3]
    xdot[r][3] = ((M+m)*m*g*L*torch.sin(x[r][1])+m*L*(-m*L*(x[r][3]**2)*torch.sin(x[r][1])-b*x[r][2])*torch.cos(x[r][1]))/((M+m)*(I+m*L*L)-m*m*L*L*torch.cos(x[r][1])**2) 
    xdot[r][2] = (-m*L*x[r][3]*x[r][3]*torch.sin(x[r][1])+m*L*xdot[r][3]*torch.cos(x[r][1])-b*x[r][2])/(M+m)
  return xdot

# x - x, theta, xdot, thetadot
def pendulumcart(x,u):
  if u[0]>10:
    u[0]=10
  elif u[0]<-10:
    u[0]=-10
  xdot = np.zeros((4,))
  xdot[0] = x[2]
  xdot[1] = x[3]
  xdot[3] = ((M+m)*m*g*L*np.sin(x[1])+m*L*(u[0]-m*L*(x[3]**2)*np.sin(x[1])-b*x[2])*np.cos(x[1]))/((M+m)*(I+m*L*L)-m*m*L*L*np.cos(x[1])**2) 
  #print(u[0]/(M+m))
  xdot[2] = (u[0]-m*L*x[3]*x[3]*np.sin(x[1])+m*L*xdot[3]*np.cos(x[1])-b*x[2])/(M+m)
  #xdot[2] = xdot[2]+u[0]/(M+m)  
  return xdot+0.2*np.random.rand(4,)
```

```python
A = np.array([[0,0,1,0],[0,0,0,1],[0,(m*m*g*L*L)/(I*(M+m)+M*m*L*L),-((I+m*L*L)*b)/(I*(M+m)+M*m*L*L),0],[0,(m*g*L*(M+m))/(I*(M+m)+M*m*L*L),(-m*L*b)/(I*(M+m)+M*m*L*L),0]])
B = np.array([[0],[0],[(I+m*L*L)/(I*(M+m)+M*m*L*L)],[(m*L)/(I*(M+m)+M*m*L*L)]])
C = np.eye(4)
D = 0
print(A)
Q = np.diag([1,1,1,1])
R = 1
Dlqr = control.lqr(A,B,Q,R)
Klqr = np.array(Dlqr[0])
Slqr = np.array(Dlqr[1])
print(Klqr)
print(Slqr)
```

```python
def lyapfuncgen(xc,lyapvector,lyapmatrix):
  a = lyapvector[0]
  b = lyapvector[1]
  c = lyapvector[2]
  d = lyapvector[3]
  e = lyapvector[4]
  f = lyapvector[5]
  g = lyapvector[6]
  h = lyapvector[7]
  i = lyapvector[8]  
  j = lyapvector[9]
  lyapfunc = (a*(1-torch.cos(b*xc[:,1]))+c*(torch.sin(d*xc[:,1]))+e*xc[:,1]+f*xc[:,0]+g*xc[:,2]+h*xc[:,3])**2\
             +i*torch.sum(torch.mul(torch.mm(xc,lyapmatrix),xc),dim=1)
  return lyapfunc
  
```

```python
def lyaptraining(dataset,Kcontrol,lyapvector,lyapmatrix,ls,max_iters,Q,R,batch_size,learning_rate,controllertrain):
  iters = 0
  costs = []
  torch.manual_seed(2)

  if not controllertrain:
    optimizer = torch.optim.Adam([lyapvector], lr=learning_rate, weight_decay=0.1)
    optimizer2 = torch.optim.Adam([lyapmatrix], lr=learning_rate, weight_decay=0.1)
  else:
    optimizer = torch.optim.Adam(Kcontrol.parameters(), lr=learning_rate, weight_decay=0.1)

  #roalearn = torch.optim.Adam([ls],lr=learning_rate)
  while iters<max_iters:
    indices = np.random.randint(low=0, high=dataset.shape[0], size=batch_size)
    x = dataset[indices,:].clone().detach().requires_grad_(False)
    #print(dataset[indices+1,:])
    x.to(device)
    xc = x.cuda()
    u = Kcontrol(xc)
    lsn = F.relu(ls.cuda())
    u = torch.clamp(u,-10,10)
    uc = u.cpu()
    xdot = pendulumcartt(xc,u)
    xdotu = torch.zeros((batch_size,4))
    xdotu[:,3] = torch.div((m*L*u[:,0]*torch.cos(xc[:,1])),((M+m)*(I+m*L*L)-m*m*L*L*torch.cos(xc[:,1])**2))
    xdotu[:,2] = (m*L*xdotu[:,3]*torch.cos(x[:,1]))/(M+m)
    xdotu[:,2] = xdotu[:,2]+uc[:,0]/(M+m)
    xdotu.to(device)
    xdot = xdot+xdotu.cuda();
    
    xnext = x.cuda()+0.01*xdot.cuda();

    xnext.to(device)
    
    lyapfunc = lyapfuncgen(xc,lyapvector.cuda(),lyapmatrix.cuda())
    lyapfuncnext = lyapfuncgen(xnext.cuda(),lyapvector.cuda(),lyapmatrix.cuda())
    #print(lyapfunc)
    #print(lyapfuncnext)
    lyapder = lyapfuncnext-lyapfunc
    lqrcost = torch.sum(torch.mul(torch.mm(x.cuda(),Q.cuda()),x.cuda()),dim=1)+torch.sum(R.cuda()*torch.mul(u,u),dim=1)
    
    cost =  (torch.div(0.5*torch.mul(torch.sign(ls.cuda()-lyapfunc)+1,F.relu(lyapder+0.9*lqrcost+0.01)),0.01*(lyapfunc+0.1))).mean()\
            #+0.1*(torch.sign(lyapder)*(ls.cuda()-lyapfunc)).mean()

    if iters%100==0:
      print(iters,"Loss=",cost.item(),"Lyapunovfunc",lyapfunc.mean().data,"Lyapunov deriv",lyapder.mean().data,"LQR cost",lqrcost.mean().data,"ls",ls)
    costs.append(cost.item())
    optimizer.zero_grad()
    optimizer2.zero_grad()
    #roalearn.zero_grad()
    cost.backward()
    optimizer.step()
    optimizer2.step()
    #roalearn.step()
    iters = iters+1
  return costs,Kcontrol,lyapvector,lyapmatrix,ls
```

```python
class controllermed(torch.nn.Module):
   def __init__(self):
    super(controllermed, self).__init__()
    torch.manual_seed(2)
    self.control1 = torch.nn.Linear(4,5,bias=False)
    self.control2 = torch.nn.Linear(5,1,bias=False)
  
   def forward(self,x):
    sigmoid = torch.nn.Tanh()
    contr1 = sigmoid(self.control1(x))
    contr = self.control2(contr1)
    return contr
```

```python
def generatedataset(startstates):
    dataset = []
    #print(startstates.shape[0])
    for i in range(startstates.shape[0]):
      #print(i)
      T = 2000
      x = startstates[i,:];
      startdataset = []
      for t in np.arange(0,T):
        xdot = pendulumcart(x,-np.dot(Klqr,x))
        x = x+0.01*xdot;
        #if(t%3==0):
        startdataset.append(list(x))            
        #print(x)
        if(np.linalg.norm(x)<0.1):
          break;
      if(np.linalg.norm(x)<0.1):
         dataset = dataset + startdataset
    return dataset  
```
