```python
import numpy as np
import scipy
import torch
from matplotlib import pyplot as plt
import torch.optim as optim
import torch.nn.functional as F
from torch.autograd import Variable 
import control
from control import matlab
from matplotlib import cm

np.random.seed(0)
dev = "cpu"  
device = torch.device(dev) 
```

```python
sysA = np.array([[1.1,-0.01,-0.01],[0.01,1,0],[0,0.01,1]])
#sysA = 0.99
sysB = np.array([[0.0199],[0],[0]])
sysC = np.array([0,1,0.5])
sysD = 0
print(np.linalg.eig(sysA))
kp = 50.0
ki = 0.1
kd = 10.0
Tf = 0.01
```

```python
refs = np.zeros((10,1000))

refs[0,0:1000] = 2*(0.1*np.sin(0.1*np.arange(0,1000))+0.3*np.sin(0.01*np.arange(0,1000)))
refs[0,0:1000] = 2*(0.4*np.sin(0.1*np.arange(0,1000))+0.5*np.sin(0.01*np.arange(0,1000)))
refs[2,0:1000] = 2*(0.4*np.sin(0.1*np.arange(0,1000))+0.7*np.sin(0.01*np.arange(0,1000)))
refs[3,0:1000] = 2*(0.8*np.sin(0.1*np.arange(0,1000))+0.4*np.sin(0.01*np.arange(0,1000)))
refs[4,0:1000] = 0.6*refs[0,:]+0.4*refs[2,:]

refs[5,100:700] = 3

refs[6,100:700] = -3

refs[7,100:700] = -2

refs[8,100:700] = 2

refs[9,:] = 0.5*refs[5,:]+0.5*refs[8,:]
```

```python
x = np.array([[0],[0],[0]])

ref = refs[5,:]

ysat = np.zeros((1000,))
uvalssat = np.zeros((1000,))
xcont = 0;
u = 0;
D = 0;
ad = Tf/(Tf+0.01)
bd = kd/(Tf+0.01)
noise = 0.01*np.random.randn(1000,1)
for i in range(1000):
   ysat[i] = np.dot(sysC,x)+np.dot(sysD,u)+noise[i];
   xcont = xcont+(ref[i]-ysat[i]);
   if(i>0):
     D = ad*D-bd*(ysat[i]-ysat[i-1]);  
     u = 0.01*ki*xcont+kp*(ref[i]-ysat[i])+D
   if(u>125):
     u=125;
   elif(u<-125):
     u=-125  
   uvalssat[i] = u;
   x = np.dot(sysA,x)+np.dot(sysB,u);

ynosat = np.zeros((1000,))
uvalsnosat = np.zeros((1000,))
xcont = 0;
u = 0;
x = np.array([[0],[0],[0]])

for i in range(1000):
   ynosat[i] = np.dot(sysC,x)+np.dot(sysD,u)+noise[i];
   #print(ysat[i])
   xcont = xcont+(ref[i]-ynosat[i]);
   if(i>0):
     D = ad*D-bd*(ynosat[i]-ynosat[i-1]);  
     u = 0.01*ki*xcont+kp*(ref[i]-ynosat[i])+D 
   uvalsnosat[i] = u;
   x = np.dot(sysA,x)+np.dot(sysB,u);

plt.figure()
plt.plot(0.01*np.arange(1000),ynosat)
plt.plot(0.01*np.arange(1000),ysat)
plt.xlabel("Time (s)")
plt.ylabel("Output")
plt.legend(["No saturation","Saturation"])
plt.show()

plt.figure()
plt.plot(0.01*np.arange(1000),uvalsnosat)
plt.plot(0.01*np.arange(1000),uvalssat)
plt.xlabel("Time (s)")
plt.ylabel("Input")
plt.legend(["No saturation","Saturation"])
plt.show()
```

```python
kppg = kp
kipg = ki
kdpg = kd
satpg = 205.0
nupdates = 10
nepisodes = 128
updatecosts = np.zeros((nupdates,))

for update in range(nupdates):
  costs = np.zeros((nepisodes,))
  kpder = np.zeros((nepisodes,))
  kider = np.zeros((nepisodes,))
  kdder = np.zeros((nepisodes,))
  satder = np.zeros((nepisodes,)) 
  for episode in range(nepisodes):
    ref = refs[5,:]
    xcont = 0;
    u = 0;
    x = np.array([[0],[0],[0]])
    sigma = 0.5
    noise = 0*np.random.randn(1000,1)
    ypg = np.zeros((1000,1))
    uvalspg = np.zeros((1000,))
    uvalspol = np.zeros((1000,))
    uvalsfinal = np.zeros((1000,))
    ad = Tf/(Tf+0.01)
    bd = kdpg/(Tf+0.01)
    D = 0
    for t in range(1000):
      ypg[t] = np.dot(sysC,x)+np.dot(sysD,u)+noise[t]
      xcont = xcont+(ref[t]-ypg[t])
      if(t>0):
        D = ad*D-bd*(ypg[t]-ypg[t-1]);
        ugain = 0.01*kipg*xcont+kppg*(ref[t]-ypg[t])+D
      else:
        ugain = 0
      upresat = np.tanh(ugain/satpg)
      uvalspol[t] = satpg*upresat
      u = uvalspol[t]+sigma*np.random.randn()
      uvalspg[t] = u
      if(u>125):
        u=125
      elif(u<-125):
        u=-125
      uvalsfinal[t] = uvalspg[t]
      x = np.dot(sysA,x)+np.dot(sysB,u)
      costs[episode] += min((ref[t]-ypg[t])**2,20)
      kpder[episode] += (ref[t]-ypg[t])*(1-upresat**2)*(uvalspg[t]-uvalspol[t])/(2*sigma)
      kider[episode] += 0.01*xcont*(1-upresat**2)*(uvalspg[t]-uvalspol[t])/(2*sigma)
      #kpder[episode] += (ref[t]-ypg[t])*(uvalspg[t]-uvalspol[t])/(2*sigma)
      #kider[episode] += 0.01*xcont*(uvalspg[t]-uvalspol[t])/(2*sigma)
      satder[episode] += (upresat-(1-upresat**2)*ugain/satpg)*(uvalspg[t]-uvalspol[t])/(2*sigma)
      kdder[episode] += -((ypg[t]-ypg[t-1])/(Tf+0.01))*(1-upresat**2)*(uvalspg[t]-uvalspol[t])/(2*sigma)
      #kdder[episode] += -((ypg[t]-ypg[t-1])/(Tf+0.01))*(uvalspg[t]-uvalspol[t])/(2*sigma)
      #kpder[episode] += -(kpt-kppg)*(ref[i]-ysat[i])/(2*sigma)
      #kider[episode] += -(kit-kipg)*0.01*xcont/(2*sigma)
    #print("Episode cost: ",costs[episode])    
  # Update kp,ki
  lr = 0.1
  updatecosts[update] = np.min(costs)
  print("Update cost",updatecosts[update])
  kppg = kppg-lr*np.mean(kpder*(costs-np.mean(costs)))
  kipg = kipg-lr*np.mean(kider*(costs-np.mean(costs)))
  satpg = satpg-lr*np.mean(satder*(costs-np.mean(costs)))
  kdpg = kdpg-lr*np.mean(kdder*(costs-np.mean(costs)))
  print(kppg,kipg,kdpg,satpg)
  #kppg = np.clip(kppg,0,200)
  #kipg = np.clip(kipg,0,200)
```

```python
# x = np.array([[0],[0],[0]])

ref = refs[5,:]

ysat = np.zeros((1000,))
uvalssat = np.zeros((1000,))
xcont = 0;
u = 0;
D = 0;
ad = Tf/(Tf+0.01)
bd = kd/(Tf+0.01)
noise = 0.01*np.random.randn(1000,1)
costsat = 0
for i in range(1000):
   ysat[i] = np.dot(sysC,x)+np.dot(sysD,u)+noise[i];
   xcont = xcont+(ref[i]-ysat[i]);
   if(i>0):
     D = ad*D-bd*(ysat[i]-ysat[i-1]);  
     u = 0.01*ki*xcont+kp*(ref[i]-ysat[i])+D
   if(u>125):
     u=125;
   elif(u<-125):
     u=-125  
   uvalssat[i] = u;
   costsat += (ref[i]-ysat[i])**2
   x = np.dot(sysA,x)+np.dot(sysB,u);
    
ypg = np.zeros((1000,))
uvalspg = np.zeros((1000,))
xcont = 0;
u = 0;
D = 0;
ad = Tf/(Tf+0.01)
bd = kdpg/(Tf+0.01)
costpg = 0
for i in range(1000):
   ypg[i] = np.dot(sysC,x)+np.dot(sysD,u)+noise[i];
   xcont = xcont+(ref[i]-ypg[i]);
   if(i>0):
     D = ad*D-bd*(ypg[i]-ypg[i-1]);  
     u = satpg*np.tanh((0.01*kipg*xcont+kppg*(ref[i]-ypg[i])+D)/satpg)
   if(u>125):
     u=125;
   elif(u<-125):
     u=-125  
   uvalspg[i] = u;
   costpg += (ref[i]-ypg[i])**2
   x = np.dot(sysA,x)+np.dot(sysB,u);    

ynosat = np.zeros((1000,))
uvalsnosat = np.zeros((1000,))
xcont = 0;
u = 0;
x = np.array([[0],[0],[0]])

for i in range(1000):
   ynosat[i] = np.dot(sysC,x)+np.dot(sysD,u)+noise[i];
   #print(ysat[i])
   xcont = xcont+(ref[i]-ynosat[i]);
   if(i>0):
     D = ad*D-bd*(ynosat[i]-ynosat[i-1]);  
     u = 0.01*ki*xcont+kp*(ref[i]-ynosat[i])+D 
   uvalsnosat[i] = u;
   x = np.dot(sysA,x)+np.dot(sysB,u);

print(costsat,costpg)    
plt.figure()
plt.plot(0.01*np.arange(1000),ynosat)
plt.plot(0.01*np.arange(1000),ysat)
plt.plot(0.01*np.arange(1000),ypg)
plt.xlabel("Time (s)")
plt.ylabel("Output")
plt.legend(["No saturation","Saturation","PG"])
plt.show()

plt.figure()
plt.plot(0.01*np.arange(1000),uvalsnosat)
plt.plot(0.01*np.arange(1000),uvalssat)
plt.plot(0.01*np.arange(1000),uvalspg)
plt.xlabel("Time (s)")
plt.ylabel("Input")
plt.legend(["No saturation","Saturation","PG"])
plt.show()
```
