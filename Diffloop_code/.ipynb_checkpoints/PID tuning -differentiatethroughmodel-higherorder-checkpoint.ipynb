```python
sysA = np.array([[1.1,-0.01,-0.01],[0.01,1,0],[0,0.01,1]])
#sysA = 0.99
sysB = np.array([[0.0199],[0],[0]])
sysC = np.array([0,1,0.5])
sysD = 0
print(np.linalg.eig(sysA))
kp = 50.0
ki = 0.1
kd = 10.0
Tf = 0.01
```

```python
x = np.array([[0],[0],[0]])

ref = np.ones((1500,))
ref[200:400] = 5;
ref[400:600] = 0;
ref[600:1000] = -3.1;
ref[1000:1400] = 2.0;
ref[1400:1500] = 5.0;

ysat = np.zeros((1500,))
uvalssat = np.zeros((1500,))
xcont = 0;
u = 0;
D = 0;
ad = Tf/(Tf+0.01)
bd = kd/(Tf+0.01)
noise = 0.05*np.random.randn(1500,1)
for i in range(1500):
   ysat[i] = np.dot(sysC,x)+np.dot(sysD,u)+noise[i];
   #print(ysat[i])
   xcont = xcont+(ref[i]-ysat[i]);
   if(i>0):
     D = ad*D-bd*(ysat[i]-ysat[i-1]);  
     u = 0.01*ki*xcont+kp*(ref[i]-ysat[i])+D
   if(u>220):
     u=220;
   elif(u<-220):
     u=-220;  
   uvalssat[i] = u;
   x = np.dot(sysA,x)+np.dot(sysB,u);
   #print(x)

ynosat = np.zeros((1500,))
uvalsnosat = np.zeros((1500,))
xcont = 0;
u = 0;
x = np.array([[0],[0],[0]])
noise = 0.05*np.random.randn(1500,1)

for i in range(1500):
   ynosat[i] = np.dot(sysC,x)+np.dot(sysD,u)+noise[i];
   #print(ysat[i])
   xcont = xcont+(ref[i]-ynosat[i]);
   if(i>0):
     D = ad*D-bd*(ynosat[i]-ynosat[i-1]);  
     u = 0.01*ki*xcont+kp*(ref[i]-ynosat[i])+D 
   uvalsnosat[i] = u;
   x = np.dot(sysA,x)+np.dot(sysB,u);
```
