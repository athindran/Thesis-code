```python
import numpy as np
import scipy
import torch
from matplotlib import pyplot as plt
import torch.optim as optim
import torch.nn.functional as F
from torch.autograd import Variable 
import control
from control import matlab
from matplotlib import cm

np.random.seed(0)
```

```python
# Define the true system model
mtrue = 0.15
btrue = 0.1
ltrue = 0.5
g = 9.81

def pendulumt(x,u):
  noise = 0
  f = []
  for r in range(0,len(x)):
    f.append([x[r][1], -(btrue/(mtrue*ltrue**2))*x[r][1]+(g/ltrue)*torch.sin(x[r][0])])
  ft = torch.tensor(f,requires_grad=False)
  return ft

def pendulum(x,u):
  if u[0]>0.8:
    u[0]=0.8
  elif u[0]<-0.8:
    u[0]=-0.8
    
  return np.array([x[1], -(btrue/(mtrue*ltrue**2))*x[1]+(g/ltrue)*np.sin(x[0])+u[0]/(mtrue*ltrue**2)])
```

```python
A = np.array([[0,1],[(g/ltrue),-1*(btrue/(mtrue*ltrue**2))]])
B = np.array([[0],[1/(mtrue*ltrue**2)]])
C = np.eye(2)
D = 0

Q = np.diag([1,0.1])
R = 0.1
Dlqr = control.lqr(A,B,Q,R)
Klqr = np.array(Dlqr[0])
Slqr = np.array(Dlqr[1])
print(Klqr)
print(Slqr)
```

```python
nx = 100
ny = 100
x = np.linspace(-3.14, 3.14, nx)
y = np.linspace(-6, 6, ny)
xv, yv = np.meshgrid(x, y)
xvnext = np.array(xv)
yvnext = np.array(yv)
verifies = np.zeros((nx,ny),dtype=np.bool)
V = np.zeros((nx,ny))
Vdot = np.zeros((nx,ny))

#plt.figure()
for i in range(nx):
    for j in range(ny):
      xvec = np.array([[xv[i,j]],[yv[i,j]]])
      xvecdot = np.array(pendulum(xvec,-np.dot(Klqr,xvec)))
      xvnext[i,j] = xvecdot[0]
      yvnext[i,j] = xvecdot[1]
      V[i,j] = np.dot(np.dot(xvec.T,Slqr),xvec)
      Vdot[i,j] = np.dot(np.dot(xvecdot.T,Slqr),xvec)  
      if V[i,j]>=0 and Vdot[i,j]<0:
        verifies[i,j] = True
fig = plt.figure()
plt.scatter(xv[verifies],yv[verifies],color='w')
plt.scatter(xv[np.logical_not(verifies)],yv[np.logical_not(verifies)],color='k')
contours = plt.contour(xv,yv,V,[0.1,0.2,0.5,1])
plt.clabel(contours, inline=1, fontsize=10)
plt.streamplot(xv,yv,xvnext,yvnext, color=('gray'), linewidth=0.5,
                  density=0.5, arrowstyle='-|>', arrowsize=1.5)
plt.xlabel('Angle (rad)')
plt.ylabel('Angular velocity (rad/s)')
plt.show()
```

```python
T = 1000
sols = np.zeros((T,3))
x = np.array([1.5,2.0]);


for t in np.arange(0,T):
    sols[t,0:2] = x;
    u = -np.dot(Klqr,x)
    #print(u)
    xdot = pendulum(x,u)
    
    lyaph = np.dot(Slqr,x)
    sols[t,2] = np.dot(x,lyaph.ravel())
    x = x+0.01*xdot;
    
plt.figure()
plt.plot(0.01*np.arange(0,T),sols[:,0])
plt.title("LQR State 0")

plt.figure()
plt.plot(0.01*np.arange(0,T),sols[:,1])
plt.title("LQR State 1")
print(sols[-1,0])
plt.figure()
plt.plot(0.01*np.arange(0,T),sols[:,2])
plt.title("lyapunov function")
print(sols[-1,2])
```

```python
class lyapNetquad(torch.nn.Module):
  def __init__(self):
    super(lyapNetquad, self).__init__()
    torch.manual_seed(2)
    self.lyaphalf1 = torch.nn.Linear(2,7,2,bias=False)
    self.ls = torch.randn(1,1) 
  
  def forward(self,x):
    lyaph = self.lyaphalf1(x)
    return lyaph,self.ls

class lyapNetmed(torch.nn.Module):
  def __init__(self):
    super(lyapNetmed, self).__init__()
    torch.manual_seed(2)
    self.lyaphalf1 = torch.nn.Linear(2,7,bias=False)
    self.lyaphalf2 = torch.nn.Linear(7,5,bias=False)
  
  def forward(self,x):
    sigmoid = torch.nn.Tanh()    
    lyaph1 = sigmoid(self.lyaphalf1(x))
    lyaph = self.lyaphalf2(lyaph1)
    return lyaph

class lyapNetlarge(torch.nn.Module):
  def __init__(self):
    super(lyapNetlarge, self).__init__()
    torch.manual_seed(2)
    self.lyaphalf1 = torch.nn.Linear(2,32,bias=False)
    self.lyaphalf2 = torch.nn.Linear(32,32,bias=False)
    self.lyaphalf3 = torch.nn.Linear(32,5,bias=False)
  
  def forward(self,x):
    sigmoid = torch.nn.Tanh()    
    lyaph1 = sigmoid(self.lyaphalf1(x))
    lyaph2 = sigmoid(self.lyaphalf2(lyaph1))
    lyaph = self.lyaphalf3(lyaph2)
    return lyaph
```

```python
def lyaptraining(dataset,Kcontrol,lyaph,ls,max_iters,Q,R,batch_size,learning_rate):
  iters = 0
  costs = []
  torch.manual_seed(2)
  
  optimizer = torch.optim.Adam((lyaph,ls), lr=learning_rate)
  
  while iters<max_iters:
    indices = np.random.randint(low=0, high=dataset.shape[0], size=batch_size)
    x = dataset[indices,:].clone().detach().requires_grad_(False)
    #lyaph,ls = net(x)
    u = Kcontrol(x)
    u = torch.clamp(u,-0.8,0.8)
    xdot = pendulumt(x,u)
    xdotu = torch.zeros((batch_size,2))
    xdotu[:,1] = u[:,0]/(mtrue*ltrue**2)
    xdot = xdot+xdotu;
    xnext = x+0.01*xdot;
    #lyaphnext,ls = net(xnext)
    
    lyapfunch = torch.mm(x,lyaph)
    lyapfunc = torch.sum(lyapfunch**2,dim=1)+0.1*torch.sum(x**2,dim=1)
    lyapfuncnh = torch.mm(xnext,lyaph)
    lyapfuncnext = torch.sum(lyapfuncnh**2,dim=1)+0.1*torch.sum(xnext**2,dim=1)
    lyapder = lyapfuncnext-lyapfunc
    lqrcost = torch.sum(torch.mul(torch.mm(x,Q),x),dim=1)+torch.sum(R*torch.mul(u,u),dim=1)
    
    cost =  (torch.div(0.5*(torch.sign(ls-lyapfunc)+1)*(F.relu(lyapder+0.9*lqrcost+1)),0.001*(lyapfunc+0.1))).mean()\
            +0.1*(torch.sign(lyapder)*(ls-lyapfunc)).mean()

    if iters%1000==0:
      print(iters,"Loss=",cost.item(),"Lyapunovfunc",lyapfunc.mean().data,"Lyapunov deriv",lyapder.mean().data,"LQR cost",lqrcost.mean().data)
    costs.append(cost.item())
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    iters = iters+1
  return costs,Kcontrol,lyaph,ls

def controltraining(dataset,Kcontrol,lyaph,ls,max_iters,Q,R,batch_size,learning_rate):
  iters = 0
  costs = []
  torch.manual_seed(2)
  
  optimizer = torch.optim.Adam((Kcontrol.parameters()), lr=learning_rate)
  
  while iters<max_iters:
    indices = np.random.randint(low=0, high=dataset.shape[0], size=batch_size)
    x = dataset[indices,:].clone().detach().requires_grad_(False)
    #lyaph,ls = net(x)
    u = Kcontrol(x)
    u = torch.clamp(u,-0.8,0.8)
    xdot = pendulumt(x,u)
    xdotu = torch.zeros((batch_size,2))
    xdotu[:,1] = u[:,0]/(mtrue*ltrue**2)
    xdot = xdot+xdotu;
    xnext = x+0.01*xdot;
    #lyaphnext,ls = net(xnext)
    
    lyapfunch = torch.mm(x,lyaph)
    lyapfunc = torch.sum(lyapfunch**2,dim=1)+0.1*torch.sum(x**2,dim=1)
    lyapfuncnh = torch.mm(xnext,lyaph)
    lyapfuncnext = torch.sum(lyapfuncnh**2,dim=1)+0.1*torch.sum(xnext**2,dim=1)
    lyapder = lyapfuncnext-lyapfunc
    lqrcost = torch.sum(torch.mul(torch.mm(x,Q),x),dim=1)+torch.sum(R*torch.mul(u,u),dim=1)
    
    cost =  (torch.div(0.5*(torch.sign(ls-lyapfunc)+1)*(F.relu(lyapder+0.9*lqrcost+1)),0.001*(lyapfunc+0.1))).mean()\
            +0.1*(torch.sign(lyapder)*(ls-lyapfunc)).mean()

    if iters%1000==0:
      print(iters,"Loss=",cost.item(),"Lyapunovfunc",lyapfunc.mean().data,"Lyapunov deriv",lyapder.mean().data,"LQR cost",lqrcost.mean().data)
    costs.append(cost.item())
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    iters = iters+1
  return costs,Kcontrol,lyaph,ls
```

```python
max_iters = 10000
learning_rate = 0.01

torch.manual_seed(2)
dataset = torch.rand([10000,2],requires_grad=False)
dataset[:,0] = 6*dataset[:,0]-3;
dataset[:,1] = 6*dataset[:,1]-3;

#print(dataset)
batch_size = 100
#lyapnetq = lyapNetquad()
Q = torch.tensor([[1,0],[0,0.1]], dtype=torch.float64)
R = torch.tensor([0.1], dtype=torch.float64)

Kcontrol = torch.nn.Linear(2,1,bias=False)
Kcontrol.weight = torch.nn.Parameter(torch.tensor(([[-3.98, -1.04]])))
lyaph = torch.randn(2,7,requires_grad=True)
ls = torch.tensor([10.0], requires_grad=True)

iters = 0
while iters<2:
  costs,Kcontrol,lyaph,ls = lyaptraining(dataset,Kcontrol,lyaph,ls,max_iters,Q.float(),R.float(),batch_size,learning_rate)
  costs,Kcontrol,lyaph,ls = controltraining(dataset,Kcontrol,lyaph,ls,max_iters,Q.float(),R.float(),batch_size,learning_rate)
  iters = iters+1
```

```python
print(lyaph)
print(ls)
print(Kcontrol.weight)
lyapnp = lyaph.detach().numpy()
Slearn = np.dot(lyapnp,lyapnp.T)
Klearn = Kcontrol.weight.detach().numpy().ravel()
print(Klearn)
print(Slearn)
```

```python
T = 1000
sols = np.zeros((T,3))
x = np.array([1.5,2.0]);


for t in np.arange(0,T):
    sols[t,0:2] = x;
    u = [np.dot(Klearn,x)]
    xdot = pendulum(x,u)
    
    lya = np.dot(Slearn,x)
    sols[t,2] = np.dot(x,lya.ravel())
    x = x+0.01*xdot;
    
plt.figure()
plt.plot(0.01*np.arange(0,T),sols[:,0])
plt.title("LQR State 0")

plt.figure()
plt.plot(0.01*np.arange(0,T),sols[:,1])
plt.title("LQR State 1")
print(sols[-1,0])
plt.figure()
plt.plot(0.01*np.arange(0,T),sols[:,2])
plt.title("lyapunov function")
print(sols[-1,2])
```
